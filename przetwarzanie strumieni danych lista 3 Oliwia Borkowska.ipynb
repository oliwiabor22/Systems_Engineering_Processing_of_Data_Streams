**Zad 2.** Przygotuj w Pythonie kod, który wyznaczy z definicji widmową gęstość mocy. WGM z definicji wyznaczana jest z zależności:

```python
# sposób nr 1 - za pomocą funkcji, z normalizacją i pełnym widmem (dodatnie i ujemne częstotliwości)

def psd_from_definition(sinus, f):
    N = len(sinus)
    fft_values = np.fft.fft(sinus) # transformacja Fouriera
    psd = (np.abs(fft_values) ** 2) / N  # normalizacja przez N
    psd = psd / f  # dodatkowa normalizacja
    freqs = np.fft.fftfreq(N, d=1/f)
    psd_shifted = np.fft.fftshift(psd)
    freqs_shifted = np.fft.fftshift(freqs)
    return freqs_shifted, psd_shifted

freqs_d, S_normalized_shifted = psd_from_definition(sinus, f)

plt.figure(figsize=(10, 5))
plt.plot(freqs_d, S_normalized_shifted, label='PSD z definicji (po normalizacji i przesunięciu)', color='blue')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Widmowa Gęstość Mocy (PSD)')
plt.title('Widmowa Gęstość Mocy z definicji (PSD)')
plt.legend()
plt.grid()
plt.show()
```

```python
# sposób nr 2 za pomocą autokorelacji
# R to funkcja autokorelacji sygnału
# S to widomowa gęstość mocy

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, fftshift

#autokorelacja
def autocorrelation(x):
    n = len(x)
    result = np.correlate(x, x, mode='full') / n  # znormalizowanie
    return result[n - 1:]
R = autocorrelation(sinus)

# transformacja Fouriera autokorelacji
S  = np.abs(fft(R))
freqs = fftfreq(len(R), d=1/f)  # Odpowiednie częstotliwości

# Przesunięcie widma dla lepszego wyglądu wykresu
S_shifted = fftshift(S)
freqs_shifted = fftshift(freqs)

# Wykres PSD
plt.figure(figsize=(10, 5))
plt.plot(freqs_shifted, S_shifted, label='PSD z definicji', color='blue')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Widmowa Gęstość Mocy (PSD)')
plt.title('Widmowa Gęstość Mocy z definicji (PSD)')
plt.legend()
plt.grid()
plt.show()
```

```python
# sposób nr 1 - za pomocą funkcji, z normalizacją i pełnym widmem (dodatnie i ujemne częstotliwości)
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square, sawtooth, chirp, unit_impulse, correlate


fs = 1000 
t = np.linspace(0, 1, 1000, endpoint=False)

def psd_from_definition(signal, fs):
    N = len(signal)
    fft_values = np.fft.fft(signal)
    psd = (np.abs(fft_values) ** 2) / (N * fs)
    freqs = np.fft.fftfreq(N, d=1/fs)
    return np.fft.fftshift(freqs), np.fft.fftshift(psd)

sinus = np.sin(2 * np.pi * 5 * t)
prostokatny = square(2 * np.pi * 5 * t)
piloksztaltny = sawtooth(2 * np.pi * 5 * t)
swiergotliwy = chirp(t, f0=1, f1=50, t1=1, method='linear')
superpozycja = sinus + prostokatny
impuls = unit_impulse(len(t), idx=t.size // 2)


signals = [sinus, prostokatny, piloksztaltny, swiergotliwy, superpozycja, impuls]
labels = ['Sinusoidalny', 'Prostokątny', 'Pilo-kształtny', 'Swiergotliwy', 'Superpozycja', 'Impuls']
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

for i, signal in enumerate(signals):
    freqs_d, S_normalized_shifted = psd_from_definition(signal, fs)
    ax = axs[i // 2, i % 2]
    ax.plot(freqs_d, S_normalized_shifted, label=f'PSD - {labels[i]}', color='blue')
    ax.set_xlabel('Częstotliwość [Hz]')
    ax.set_ylabel('Widmowa Gęstość Mocy (PSD)')
    ax.set_title(f'Widmowa Gęstość Mocy - {labels[i]}')
    ax.grid()
    ax.legend()
plt.suptitle("Widmowa gęstość mocy (PSD) z definicji dla wszystkich sygnałów", fontsize=20, style ='italic')
plt.tight_layout()
plt.show()
```
