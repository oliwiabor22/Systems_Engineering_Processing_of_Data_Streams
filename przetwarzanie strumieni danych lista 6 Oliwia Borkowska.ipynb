**Zad. 1** Przygotuj w Pythonie kod, który przedstawi na wykresie okno Hamminga, Hanna, Blackmana oraz Dirichleta oraz ich widma amplitudowe (pakiet SciPy).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows
from scipy.fft import fft, fftshift
from ipywidgets import interact, IntSlider

def okno_i_widmo(window_type, N):
    if window_type == "Hamming":
        window = windows.hamming(N)
    elif window_type == "Hann":
        window = windows.hann(N)
    elif window_type == "Triangle":
        window = windows.triang(N)
    elif window_type == "Blackman":
        window = windows.blackman(N)
    elif window_type == "Dirichlet":
        n = np.arange(-(N-1)/2, (N-1)/2 + 1)
        window = np.sinc(n / (N/2))
    
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(window, color="magenta")
    plt.title(f"Okno {window_type}")
    plt.xlabel("Próbka")
    plt.ylabel("Amplituda")

    plt.subplot(1, 2, 2)
    A = fft(window, 2048) / (len(window)/2.0)
    freq = np.linspace(-0.5, 0.5, len(A))
    response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
    plt.plot(freq, response)
    plt.axis([-0.5, 0.5, -120, 0])
    plt.title(f"Widmo amplitudowe okna {window_type}")
    plt.xlabel("Znormalizowana częstotliwość")
    plt.ylabel("Znormalizowana amplituda [dB]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_hamming(N):
    okno_i_widmo("Hamming", N)

def plot_hann(N):
   okno_i_widmo("Hann", N)
    
def plot_triangle(N):
    okno_i_widmo("Triangle", N)

def plot_blackman(N):
    okno_i_widmo("Blackman", N)

def plot_dirichlet(N):
    okno_i_widmo("Dirichlet", N)

print("Okno Hamminga:")
interact(
    plot_hamming,
    N=IntSlider(value=64, min=16, max=256, step=8, description='Rozmiar okna:')
)

print("\nOkno Hanna:")
interact(
    plot_hann,
    N=IntSlider(value=64, min=16, max=256, step=8, description='Rozmiar okna:')
)

print("\nOkno Triangle:")
interact(
    plot_triangle,
    N=IntSlider(value=64, min=16, max=256, step=8, description='Rozmiar okna:')
)

print("\nOkno Blackmana:")
interact(
    plot_blackman,
    N=IntSlider(value=64, min=16, max=256, step=8, description='Rozmiar okna:')
)

print("\nOkno Dirichleta:")
interact(
    plot_dirichlet,
    N=IntSlider(value=64, min=16, max=256, step=8, description='Rozmiar okna:')
)
```

**Zad. 2** Przygotuj w Pythonie kod, który wyznaczy widmo sygnału sinusoidalnego o trzech różnych częstotliwościach przy zastosowanych oknach: Hamminga, Hanna, Blackmana oraz Dirichleta

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows
from scipy.fft import fft, fftshift
from ipywidgets import interact, IntSlider, FloatSlider, Dropdown

def generuj_sygnal(freq1, freq2, freq3, fs, N): # generowanie sygnału złozonego z3 sinusoid o podanych f
    t = np.arange(N) / fs
    signal = np.sin(2 * np.pi * freq1 * t) + np.sin(2 * np.pi * freq2 * t) + np.sin(2 * np.pi * freq3 * t)
    return signal

def get_window(window_type, N):

    if window_type == "Hamming":
        return windows.hamming(N)
    elif window_type == "Hann":
        return windows.hann(N)
    elif window_type == "Blackman":
        return windows.blackman(N)
    elif window_type == "Dirichlet":
        # Okno Dirichleta (prostokątne)
        return np.ones(N)
    elif window_type == "Triangular":
        # Okno trójkątne (Bartletta)
        return windows.bartlett(N)
    return np.ones(N)

def analiza_sygnalu_okna(window_type, freq1, freq2, freq3, N, fs=1000):
    signal = (freq1, freq2, freq3, fs, N)
    window = get_window(window_type, N)
    windowed_signal = signal[:N] * window
    
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(windowed_signal)
    plt.title(f"Sygnał z oknem {window_type}")
    plt.xlabel("Próbka")
    plt.ylabel("Amplituda")
    plt.grid(True)
    
    plt.subplot(1, 2, 2)
    A = fft(windowed_signal, 2048) / (N/2.0)
    freq = np.linspace(-fs/2, fs/2, len(A))
    response = 20 * np.log10(np.abs(fftshift(A / abs(A).max() + 1e-10)))
    plt.plot(freq, response)
    plt.axis([-fs/2, fs/2, -80, 0])
    plt.title(f"Widmo sygnału (okno {window_type})")
    plt.xlabel("Częstotliwość [Hz]")
    plt.ylabel("Amplituda [dB]")
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def porownanie_okien(freq1, freq2, freq3, N, fs=1000):
    signal = generuj_sygnal(freq1, freq2, freq3, fs, N)
    window_types = ['Hamming', 'Hann', 'Blackman', 'Dirichlet', 'Triangular']
    
    plt.figure(figsize=(12, 10))
    
    # okna
    plt.subplot(2, 1, 1)
    for w_type in window_types:
        window = get_window(w_type, N)
        plt.plot(window, label=f'{w_type}')
    plt.title("Porównanie okien")
    plt.xlabel("Próbka")
    plt.ylabel("Amplituda")
    plt.legend()
    plt.grid(True)
    
    # widma
    plt.subplot(2, 1, 2)
    for w_type in window_types:
        window = get_window(w_type, N)
        windowed_signal = signal[:N] * window
        A = fft(windowed_signal, 2048) / (N/2.0)
        freq = np.linspace(-fs/2, fs/2, len(A))
        response = 20 * np.log10(np.abs(fftshift(A / abs(A).max() + 1e-10)))
        plt.plot(freq, response, label=f'{w_type}')
    
    plt.axis([-fs/2, fs/2, -80, 0])
    plt.title("Porównanie widm dla różnych okien")
    plt.xlabel("Częstotliwość [Hz]")
    plt.ylabel("Amplituda [dB]")
    plt.legend()
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()


print("Analiza pojedynczego okna:")
interact(
    analiza_sygnalu_okna,
    window_type=Dropdown(options=['Hamming', 'Hann', 'Blackman', 'Dirichlet', 'Triangular'], value='Hamming', description='Okno:'),
    freq1=FloatSlider(value=50, min=10, max=200, step=5, description='Częst. 1 [Hz]:'),
    freq2=FloatSlider(value=100, min=10, max=200, step=5, description='Częst. 2 [Hz]:'),
    freq3=FloatSlider(value=150, min=10, max=200, step=5, description='Częst. 3 [Hz]:'),
    N=IntSlider(value=256, min=64, max=1024, step=64, description='Rozmiar okna:'),
)


print("\nPorównanie wszystkich okien:")
interact(
    porownanie_okien,
    freq1=FloatSlider(value=50, min=10, max=200, step=5, description='Częst. 1 [Hz]:'),
    freq2=FloatSlider(value=100, min=10, max=200, step=5, description='Częst. 2 [Hz]:'),
    freq3=FloatSlider(value=150, min=10, max=200, step=5, description='Częst. 3 [Hz]:'),
    N=IntSlider(value=256, min=64, max=1024, step=64, description='Rozmiar okna:'),
)
```

**Zad. 3** Przygotuj w Pythonie kod, który wyznaczy widmo sygnału sinusoidalnego z wykorzystaniem Szybkiej Transformaty Fouriera (np. pakiet SciPy oferuje funkcję fft)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift, fftfreq
from ipywidgets import interact, FloatSlider, IntSlider, Layout

def analiza_sygnalu_sinus(frequency, amplitude, phase, fs, N, window_type):

    t = np.arange(N) / fs
    signal = amplitude * np.sin(2 * np.pi * frequency * t + phase)
    
    # okno
    if window_type == 'Prostokątne':
        window = np.ones(N)
    elif window_type == 'Hamminga':
        window = np.hamming(N)
    elif window_type == 'Hanna':
        window = np.hanning(N)
    elif window_type == 'Blackmana':
        window = np.blackman(N)
    elif window_type == 'Triangle':
        window = np.triang(N)
    
    windowed_signal = signal * window
    
    # FFT
    signal_fft = fft(windowed_signal)
    signal_fft_shifted = fftshift(signal_fft)

    freq = fftshift(fftfreq(N, 1/fs))
    
    magnitude_db = 20 * np.log10(np.abs(signal_fft_shifted) / np.max(np.abs(signal_fft_shifted)) + 1e-10) # widmo amplitudowe w dB
    
    plt.figure(figsize=(14, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(t, signal, label='Sygnał oryginalny',  color='purple')
    plt.plot(t, windowed_signal, label=f'Sygnał z oknem {window_type}')
    plt.title(f'Sygnał sinusoidalny ({frequency} Hz)')
    plt.xlabel('Czas [s]')
    plt.ylabel('Amplituda')
    plt.legend()
    plt.grid(True)
    

    plt.subplot(1, 2, 2)
    plt.plot(freq, magnitude_db,  color='magenta')
    plt.title('Widmo amplitudowe sygnału')
    plt.xlabel('Częstotliwość [Hz]')
    plt.ylabel('Amplituda [dB]')
    plt.xlim(-fs/2, fs/2)
    plt.ylim(-100, 5)
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

interact(analiza_sygnalu_sinus,
    frequency=FloatSlider(min=1, max=500, step=1, value=50, description='Częstotliwość [Hz]:', 
                        layout=Layout(width='500px')),
    amplitude=FloatSlider(min=0.1, max=5, step=0.1, value=1, description='Amplituda:', 
                        layout=Layout(width='500px')),
    phase=FloatSlider(min=0, max=2*np.pi, step=0.1, value=0, description='Faza [rad]:', 
                     layout=Layout(width='500px')),
    fs=IntSlider(min=500, max=10000, step=100, value=1000, description='Częst. próbkowania [Hz]:', 
                layout=Layout(width='500px')),
    N=IntSlider(min=64, max=4096, step=64, value=1024, description='Liczba próbek:', 
               layout=Layout(width='500px')),
    window_type=(['Prostokątne', 'Hamminga', 'Hanna', 'Blackmana', 'Triangle']))
```

**Zad. 4** Przygotuj w Pythonie kod, który wyznaczy widmo dowolnego sygnału załadowanego z pliku np. csv.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fft import fft, fftshift, fftfreq
from scipy.signal import windows
from ipywidgets import interact, IntSlider, Dropdown, Text, FloatSlider, Layout

def analyze_signal(file_name='arrhytmia.csv', signal_column='MLII', fs=360, 
                  window_type='Hann', fft_size=1024, log_scale=True):
    try:
        df = pd.read_csv(file_name, sep=';') # wczytanie danych
        
        
        if signal_column not in df.columns:
            signal_column = df.columns[2]
            
        signal = df[signal_column].values
        time = df['time_ms'].values / 1000
        N = len(signal)
        
        plt.figure(figsize=(12, 4))
        plt.plot(time, signal, color='pink')
        plt.title(f'Surowy sygnał: {signal_column}')
        plt.xlabel('Czas [s]')
        plt.ylabel('Amplituda')
        plt.grid(True)
        plt.show()

        # sygnał z oknem i widmo
        window = windows.get_window(window_type, N)
        windowed_signal = signal * window
        
        # FFT
        freq = fftfreq(fft_size, 1/fs)
        fft_values = fft(windowed_signal, fft_size)
        magnitude = np.abs(fftshift(fft_values))
        magnitude_db = 20 * np.log10(magnitude / np.max(magnitude) + 1e-10)
        
    
        plt.figure(figsize=(14, 6))
        
        plt.subplot(1, 2, 1)
        plt.plot(time, windowed_signal, color='magenta')
        plt.title(f'Sygnał z oknem {window_type}')
        plt.xlabel('Czas [s]')
        plt.ylabel('Amplituda')
        plt.grid(True)
        
        plt.subplot(1, 2, 2) # widmo
        if log_scale:
            plt.plot(fftshift(freq), magnitude_db)
            plt.ylim(-120, 5)
            plt.ylabel('Amplituda [dB]')
        else:
            plt.plot(fftshift(freq), magnitude/np.max(magnitude))
            plt.ylabel('Amplituda znormalizowana')
            
        plt.title('Widmo częstotliwościowe')
        plt.xlabel('Częstotliwość [Hz]')
        plt.xlim(-fs/2, fs/2)
        plt.grid(True)
        
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Błąd: {str(e)}")

interact(
    analyze_signal,
    file_name=Text(value='arrhytmia.csv', description='Plik:', layout=Layout(width='300px')),
    signal_column=Dropdown(options=['MLII', 'V5'], value='MLII', description='Kolumna:'),
    fs=IntSlider(value=360, min=100, max=1000, step=10, description='Częst. próbkowania:'),
    window_type=Dropdown(options=['hann', 'hamming', 'blackman', 'boxcar', 'bartlett'], 
                        value='hann', description='Okno:'),
    fft_size=IntSlider(value=1024, min=256, max=4096, step=256, description='Rozmiar FFT:'),
    log_scale=Dropdown(options=[True, False], value=True, description='Skala log:')
)
```
